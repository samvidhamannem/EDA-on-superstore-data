```python
data1=data.drop(['Postal Code'],axis=1)
```

```python
data1.head()
```

# Plotting heat map

```python
corelation=data1.corr()
```

```python
sb.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

# Plot pair plots

```python
sb.pairplot(corelation)
```

```python
sb.pairplot(data1,palette='coolwarm')
```

# Relational plot-Sales x Profit

```python
sb.relplot(x='Sales',y='Profit',hue='Category',data=data1)
```

# Scatter plot-Overall discount x Profit

```python
# We plot this to see the relationship between discount and profit
plt.scatter(x=data1['Discount'],y=data1['Profit'])
plt.xlabel("Overall Discount")
plt.ylabel("Profit")
plt.show()
```

# Count plots

```python
#total count of different ship modes
sb.set_style('whitegrid')
sb.countplot(x='Ship Mode',data=data1,palette='BrBG')
```

```python
# Compare shipmode and region 
sb.set_style('whitegrid')
sb.countplot(x='Ship Mode',hue='Region',data=data1,palette='cubehelix')
```

```python
#compare segmment and category
sb.set_style('whitegrid')
sb.countplot(x='Segment',hue='Category',data=data1,palette='cubehelix')
```

```python
#compare region and category
sb.countplot(x='Region',hue='Category',data=data1,palette='cubehelix')
```
